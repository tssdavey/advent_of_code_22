def num_empty_positions(sensors):
    # initialize a list of length equal to the length of the tenth row
    # with each element set to True
    empty_positions = [True] * 100

    # iterate over the sensors
    for sensor in sensors:
        # get the x and y position of the sensor
        x, y = sensor

        # if the sensor is not on the tenth row, skip it
        if y != 10:
            continue

        # calculate the manhattan distance from the sensor to its closest beacon
        closest_beacon = (25, 17)
        distance = abs(x - closest_beacon[0]) + abs(y - closest_beacon[1])

        # iterate over all positions in the tenth row
        for i in range(100):
            # calculate the manhattan distance from the current position
            # to the sensor's closest beacon
            distance_to_beacon = abs(i - closest_beacon[0]) + abs(y - closest_beacon[1])

            # if the distance to the beacon is less than or equal to the distance
            # from the sensor to the beacon, the position cannot be empty
            if distance_to_beacon <= distance:
                empty_positions[i] = False

    # return the number of empty positions in the tenth row
    return sum(empty_positions)
