```python
def solve(packet1, packet2):
    # If both values are integers, compare them directly.
    #print('<-----')
    #print(packet1,packet2)
    if isinstance(packet1, int) and isinstance(packet2, int):
        #print('both ints')
        #print('returning',packet1 <= packet2)
        #print('----->')
        if packet1 < packet2: return 't'
        elif packet1 == packet2: return 'e'
        elif packet1 > packet2: return 'f'

    # If exactly one value is an integer, convert it to a list containing that integer.
    if isinstance(packet1, int):
        packet1 = [packet1]
    elif isinstance(packet2, int):
        packet2 = [packet2]

    # Compare the two values by comparing each item in the lists in order.
    for item1, item2 in zip(packet1, packet2):
        # If the two items are in the right order, continue to the next item.
        ##print('items',item1,item2)
        res = solve(item1, item2)
        #print('res',res)
        if res == 'e':
            #print('continuing')
            continue
        elif res == 't':
            #print('returning true')
            #print('----->')
            return 't'
        # If the two items are not in the right order, return False.
        elif res == 'f':
            #print('returning false')
            #print('----->')
            return 'f'
    # If all the items in the lists have been compared and no decision has been made,
    # return whether the left list is shorter than the right list.
    #print('returning length')
    #print('----->')
    #print(len(packet1),len(packet2))
    if len(packet1) < len(packet2): return 't'
    elif len(packet1) == len(packet2): return 'e'
    elif len(packet1) > len(packet2): return 'f'


assert(solve([1,1,3,1,1],[1,1,5,1,1]) == 't')
assert(solve([[1],[2,3,4]],[[1],4]) == 't')
assert(solve([9],[[8,7,6]]) == 'f')
assert(solve([[4,4],4,4],[[4,4],4,4,4]) == 't')
assert(solve([7,7,7,7],[7,7,7]) == 'f')
assert(solve([],[3]) == 't')
assert(solve([[[]]],[[]]) == 'f')
assert(solve([1,[2,[3,[4,[5,6,7]]]],8,9],[1,[2,[3,[4,[5,6,0]]]],8,9]) == 'f')
```

```python
#this is what openai gave me in response to this prompt
"""Can you write a python function that solves this problem. The function should take a pair of packets as input, and return wether it is in the right order or not. Return only the code no explanation
Your list consists of pairs of packets; pairs are separated by a blank line. You need to identify how many pairs of packets are in the right order.
[1,[2,[3,[4,[5,6,7]]]],8,9]
[1,[2,[3,[4,[5,6,0]]]],8,9]
Packet data consists of lists and integers. Each list starts with [, ends with ], and contains zero or more comma-separated values (either integers or other lists). Each packet is always a list and appears on its own line.

When comparing two values, the first value is called left and the second value is called right. Then:

If both values are integers, the lower integer should come first. If the left integer is lower than the right integer, the inputs are in the right order. If the left integer is higher than the right integer, the inputs are not in the right order. Otherwise, the inputs are the same integer; continue checking the next part of the input.
If both values are lists, compare the first value of each list, then the second value, and so on. If the left list runs out of items first, the inputs are in the right order. If the right list runs out of items first, the inputs are not in the right order. If the lists are the same length and no comparison makes a decision about the order, continue checking the next part of the input.
If exactly one value is an integer, convert the integer to a list which contains that integer as its only value, then retry the comparison. For example, if comparing [0,0,0] and 2, convert the right value to [2] (a list containing 2); the result is then found by instead comparing [0,0,0] and [2]."""
def in_order(packet1, packet2):
    # If both values are integers, compare them directly.
    if isinstance(packet1, int) and isinstance(packet2, int):
        return packet1 < packet2

    # If exactly one value is an integer, convert it to a list containing that integer.
    if isinstance(packet1, int):
        packet1 = [packet1]
    elif isinstance(packet2, int):
        packet2 = [packet2]

    # Compare the two values by comparing each item in the lists in order.
    for item1, item2 in zip(packet1, packet2):
        # If the two items are in the right order, continue to the next item.
        if in_order(item1, item2):
            continue
        # If the two items are not in the right order, return False.
        elif not in_order(item1, item2):
            return False
    # If all the items in the lists have been compared and no decision has been made,
    # return whether the left list is shorter than the right list.
    return len(packet1) < len(packet2)
```
